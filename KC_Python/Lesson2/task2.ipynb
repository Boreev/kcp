```python
import pandas as pd
```

```python
path_to_file = '2_taxi_nyc.csv'
```

```python
taxi = pd.read_csv(path_to_file)
```

```python
taxi = taxi.rename(columns={'pcp 01': 'pcp_01', 
                   'pcp 06': 'pcp_06', 
                   'pcp 24': 'pcp_24'})
```

```python
taxi.pickups.sum()
```

```python
borough_pickups = taxi.groupby('borough') \
    .aggregate({'pickups': 'sum'}) \
    .sort_values('pickups', ascending=False)
```

```python
borough_pickups
```

```python
min_pickups = borough_pickups.idxmin()
```

```python
min_pickups
```

```python
min_pickups = taxi.groupby('borough') \
    .aggregate({'pickups': 'sum'}) \
    .idxmin()
```

```python
min_pickups
```

```python
taxi.groupby(['borough', 'hday'], as_index=False) \
    .aggregate({'pickups': 'mean'}) \
    .sort_values('pickups', ascending=False)
```

```python
pickups_by_mon_bor = taxi.groupby(['pickup_month', 'borough'], as_index=False) \
    .aggregate({'pickups': 'sum'}) \
    .sort_values('pickups', ascending=False)
```

```python
# формула для перевода
taxi['temp_C'] = (taxi['temp'] - 32) * 5 / 9
taxi['temp_C'][:5]
```

```python
def temp_to_celcius(column_name):
    result = (column_name - 32) * 5 / 9
    return result
```

```python
taxi['temp_C'] = temp_to_celcius(taxi['temp'])
taxi['temp_C'][:5]
```
