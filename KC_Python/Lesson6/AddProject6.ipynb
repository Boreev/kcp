```python
gff = pd.read_csv(path_to_data, sep="\t")
```

```python
gff.attributes = gff.attributes.str.split(';')
```

```python
gff.head()
```

```python
def parent_search(stroka):
    for i in stroka:
        if 'Parent' in i:
            return i
```

```python
gff['parent'] = gff.attributes.apply(parent_search)
```

```python
gff.head()
```

```python
gff.parent = gff.parent.str.split('=')
```

```python
gff.head()
```

```python
def XM_search(stroka):
    for i in stroka:
        if 'XM' in i:
            return i
```

```python
gff['XM'] = gff.parent.apply(XM_search)
```

```python
gff.head()
```

```python
gff['XM'] = gff['XM'].str.strip(',')
```

```python
gff.groupby('XM', as_index=False) \
    .agg({'end': 'count'})
```
