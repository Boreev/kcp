```python
r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?departmentId=6&q=cat')
cats = r.json()
```

```python
cat_ids = cats['objectIDs']
```

```python
import pandas as pd 
df = pd.DataFrame()


for i in cat_ids[:1000]:
        path = f'https://collectionapi.metmuseum.org/public/collection/v1/objects/{i}'
        r_test = requests.get(path)
        obj_test = pd.DataFrame([r_test.json()])
        df = df.append(obj_test)

 
```

```python
df = df[['objectID', 'title', 'artistDisplayName', 'department', 'objectBeginDate', 'objectEndDate', 'period', 'objectName', 'culture']]
df
```

```python
period = df.groupby('period', as_index=False) \
    .agg({'objectID': 'count'}) \
    .sort_values('objectID', ascending=False) \

period
```
